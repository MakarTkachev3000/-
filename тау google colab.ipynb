```python
import numpy as np
from scipy.special import erf
import matplotlib.pyplot as plt

# параметры
C = 1.0          # уровень насыщения
J2 = 0.5         # интеграл (задаём как константу)
tol = 1e-6

# функция k^(2)
def k2(Dx):
    return (2*C/np.sqrt(2*np.pi*Dx))*np.exp(-C**2/(2*Dx)) + erf(C/np.sqrt(2*Dx))

# итерационное решение
Dx = 1.0  # начальное приближение

for _ in range(100):
    k = k2(Dx)
    Dx_new = (k**2) * J2
    if abs(Dx_new - Dx) < tol:
        break
    Dx = Dx_new

# итог
k_final = k2(Dx)

print("Dx =", Dx)
print("k^(2) =", k_final)

# график зависимости k^(2) от Dx
Dx_vals = np.linspace(0.01, 5, 200)
k_vals = k2(Dx_vals)

plt.plot(Dx_vals, k_vals)
plt.xlabel("Dx")
plt.ylabel("k^(2)")
plt.title("Зависимость коэффициента статистической линеаризации")
plt.grid()
plt.show()
```
